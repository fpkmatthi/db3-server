```python
# import the library 
import pandas as pd
iris = pd.read_csv('/vagrant/provisioning/files/Data_analysis/machine_learning/input-files/iris.csv',sep=',')
iris.head()
```

```python
iris.groupby('class').count()
```

```python
print(iris.head())
```

```python
import sklearn
from sklearn.model_selection import train_test_split
X = iris.drop('class',axis=1)
y = iris['class']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30)
```

```python
print(X_train.count())
print(X_test.count())
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
```

```python
y_test2 = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_test2)
```

```python
results = pd.DataFrame({'true':y_test,'estimated':y_test2})
results
```

```python
import numpy as np
results['TP'] = np.where((results['true'] == 1) & (results['estimated'] == 1),1,0)
results['TN'] = np.where((results['true'] == 0) & (results['estimated'] == 0),1,0)
results['FP'] = np.where((results['true'] == 0) & (results['estimated'] == 1),1,0)
results['FN'] = np.where((results['true'] == 1) & (results['estimated'] == 0),1,0)
```
